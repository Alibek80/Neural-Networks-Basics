Для целей домашнего задания сделал обзор алгоритма CenterNet, описание которого было опубликовано в апреле 2019 года на сайте arxiv.org: https://arxiv.org/abs/1904.07850

Основное его отличие от ранее имевшихся моделей заключается в том, что поначалу находится центр объекта (как точка), далее определяются границы прямоугольника объекта. 

Алгоритм может использовать в качестве backbone конволюционных слоев такие модели, как ResNet, DLA, Hourglass 

При использовании DLA-34 backbone на датасете COCO достигается скорость более чем в 2 раза больше, чем YOLOv3 при точности выше на 4,4%

Основные преимущества алгоритма: скорость, простота, возможность использовать в 3D
